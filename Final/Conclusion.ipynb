## Winner of the Competition

```python
import matplotlib.pyplot as plt
neighborhoods = ["Lincoln-Lemington-Belmar", "Highland Park", "Central Business District", "Strip District"]
points = [5,7,8,10]
plt.figure(figsize=(10,6))
plt.bar(neighborhoods, points, color='blue')
plt.title("Neighborhood Safety Points")
plt.xlabel("Neighborhood")
plt.ylabel("Points")
plt.ylim(0, max(points)+2)
plt.xticks(rotation=20, ha='right')
for i, v in enumerate(points):
    plt.text(i, v +0.2, str(v), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
