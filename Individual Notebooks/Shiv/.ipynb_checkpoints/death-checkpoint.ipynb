### Import Statements

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

### Isolate and Sort Dataset

```python
death_age_dataset = pd.read_csv("median_age_of_death.csv", index_col="_id")
neighborhood_ages = death_age_dataset[['NEIGHBORHOOD', 'TOTAL MD AGE AT DEATH']]
neighborhood_ages = neighborhood_ages[:-2] # Last two rows are extraneous
descending_ages = neighborhood_ages.sort_values(by='TOTAL MD AGE AT DEATH', ascending=False)
descending_ages.reset_index(drop = True, inplace = True)
```

## Sorted Ranking of Neighborhoods

```python
print(descending_ages.to_string())
```

### Clean Up Empty Data for Plotting

```python
ascending_ages = descending_ages.sort_values(by='TOTAL MD AGE AT DEATH')

nan_value = float("NaN")
ages_to_plot = ascending_ages.replace("", nan_value)
ages_to_plot.replace(0, nan_value, inplace=True)
ages_to_plot.dropna(subset=['TOTAL MD AGE AT DEATH'], axis=0, inplace=True)
```

### Visual Representation

```python
graph = ages_to_plot.plot(kind='barh', x="NEIGHBORHOOD", figsize=(100,100), fontsize=70)
graph.set_title("Age of Death in Pittsburgh Neighborhoods", fontsize=100)
graph.set_xlabel("Median Age of Death", fontsize=100)
graph.set_ylabel("Neighborhood", fontsize=100)
graph.legend().set_visible(False)
plt.show()
```
